3.	Detect outliers using:
o	IQR method
o	Z-score method


### Outlier Detection using IQR Method
The Interquartile Range (IQR) method identifies outliers as data points that fall below Q1 - 1.5*IQR or above Q3 + 1.5*IQR.

```python
Q1 = df['daily_screen_time_hours'].quantile(0.25)
Q3 = df['daily_screen_time_hours'].quantile(0.75)
IQR = Q3 - Q1

lower_bound_iqr = Q1 - 1.5 * IQR
upper_bound_iqr = Q3 + 1.5 * IQR

outliers_iqr = df[(df['daily_screen_time_hours'] < lower_bound_iqr) | (df['daily_screen_time_hours'] > upper_bound_iqr)]

print(f"Outliers using IQR method for 'daily_screen_time_hours' ({len(outliers_iqr)} instances):")
display(outliers_iqr[['daily_screen_time_hours']].head())
```

```python
from scipy.stats import zscore

df['daily_screen_time_hours_zscore'] = zscore(df['daily_screen_time_hours'])

threshold_zscore = 3

outliers_zscore = df[(df['daily_screen_time_hours_zscore'] > threshold_zscore) | (df['daily_screen_time_hours_zscore'] < -threshold_zscore)]

print(f"Outliers using Z-score method (threshold={threshold_zscore}) for 'daily_screen_time_hours' ({len(outliers_zscore)} instances):")
display(outliers_zscore[['daily_screen_time_hours', 'daily_screen_time_hours_zscore']].head())
```

4.	Visualize using:
o	Box plots
o	Histograms


```python
numerical_features = ['daily_screen_time_hours', 'notifications_per_day', 'app_opens_per_day', 'weekend_screen_time']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_features):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_features):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

The histograms and skewness calculation helped in the identification of skewed distributions. The box plots, along with the IQR and Z-score methods, were used for the detection of extreme values (outliers). While no statistical outliers were found for 'daily_screen_time_hours' with the chosen thresholds, the methods were applied and visualized.

## Assignment 4: Regression Models for House Price Prediction



### 1. Load USA Housing dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Load the dataset
try:
    df_housing = pd.read_csv('usa-data.zip.csv')
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("USA_Housing.csv not found. Please upload the file.")
    # Create a dummy DataFrame for demonstration if file not found
    data = {
        'Avg. Area Income': np.random.rand(100) * 100000,
        'Avg. Area House Age': np.random.rand(100) * 5 + 4,
        'Avg. Area Number of Rooms': np.random.rand(100) * 3 + 5,
        'Avg. Area Number of Bedrooms': np.random.rand(100) * 2 + 2,
        'Area Population': np.random.rand(100) * 20000 + 10000,
        'Price': np.random.rand(100) * 1000000 + 200000,
        'Address': [f'Address {i}' for i in range(100)]
    }
    df_housing = pd.DataFrame(data)


# Display the first 5 rows and info
display(df_housing.head())
df_housing.info()
```

### 2. Preprocess data (handle missing values, scaling).

```python

print("\nMissing values before handling:\n", df_housing.isnull().sum())


df_housing = df_housing.drop('Address', axis=1)


X = df_housing.drop('Price', axis=1)
y = df_housing['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("\nData preprocessing complete. Scaled features ready for model training.")
```

### 3. Train models: Linear Regression, Polynomial Regression, SVR



#### Linear Regression

```python
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)
y_pred_lr = lr_model.predict(X_test_scaled)

mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

print("Linear Regression Metrics:")
print(f"  MAE: {mae_lr:.2f}")
print(f"  MSE: {mse_lr:.2f}")
print(f"  R2 Score: {r2_lr:.2f}")
```

#### Polynomial Regression (Degree 2)

```python

poly_transformer = PolynomialFeatures(degree=2)
X_train_poly = poly_transformer.fit_transform(X_train_scaled)
X_test_poly = poly_transformer.transform(X_test_scaled)

poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)
y_pred_poly = poly_model.predict(X_test_poly)

mae_poly = mean_absolute_error(y_test, y_pred_poly)
mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)

print("Polynomial Regression (Degree 2) Metrics:")
print(f"  MAE: {mae_poly:.2f}")
print(f"  MSE: {mse_poly:.2f}")
print(f"  R2 Score: {r2_poly:.2f}")
```

#### Support Vector Regression (SVR)

```python

svr_model = SVR()
svr_model.fit(X_train_scaled, y_train)
y_pred_svr = svr_model.predict(X_test_scaled)

mae_svr = mean_absolute_error(y_test, y_pred_svr)
mse_svr = mean_squared_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)

print("Support Vector Regression (SVR) Metrics:")
print(f"  MAE: {mae_svr:.2f}")
print(f"  MSE: {mse_svr:.2f}")
print(f"  R2 Score: {r2_svr:.2f}")
```

### 5. Compare results.

```python
results = pd.DataFrame({
    'Model': ['Linear Regression', 'Polynomial Regression', 'SVR'],
    'MAE': [mae_lr, mae_poly, mae_svr],
    'MSE': [mse_lr, mse_poly, mse_svr],
    'R2 Score': [r2_lr, r2_poly, r2_svr]
})

display(results.sort_values(by='R2 Score', ascending=False))

print("\nBased on the R2 Score (higher is better), the best performing model is:")
print(results.loc[results['R2 Score'].idxmax()])
```

## Assignment 5: Classification Using Iris Dataset



### 1. Load Iris dataset from `sklearn.datasets` and 2. Split into training/testing sets.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


iris = load_iris()
X = iris.data
y = iris.target


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Iris dataset loaded and split into training/testing sets. Features scaled.")
```

### 3. Train all classifiers.

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

nb_model = GaussianNB()
lr_model = LogisticRegression(max_iter=200, random_state=42) # Increased max_iter for convergence
svm_model = SVC(random_state=42)
dt_model = DecisionTreeClassifier(random_state=42)

nb_model.fit(X_train_scaled, y_train)

lr_model.fit(X_train_scaled, y_train)

# Train SVM
svm_model.fit(X_train_scaled, y_train)

# Train Decision Tree
dt_model.fit(X_train_scaled, y_train)

print("All classifiers trained successfully.")
```

### 4. Evaluate using Accuracy, Confusion Matrix, and Classification Report.

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def evaluate_model(model, X_test_scaled, y_test, model_name):
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred, target_names=iris.target_names)

    print(f"\n--- {model_name} ---")
    print(f"Accuracy: {accuracy:.2f}")
    print("Confusion Matrix:")
    print(cm)
    print("Classification Report:")
    print(report)

# Evaluate Naive Bayes
evaluate_model(nb_model, X_test_scaled, y_test, "Naive Bayes")

# Evaluate Logistic Regression
evaluate_model(lr_model, X_test_scaled, y_test, "Logistic Regression")

# Evaluate SVM
evaluate_model(svm_model, X_test_scaled, y_test, "Support Vector Machine")

# Evaluate Decision Tree
evaluate_model(dt_model, X_test_scaled, y_test, "Decision Tree")
```

Regression models for house price prediction using dummy data, which resulted in poor R² scores for all models. It also included classification models on the Iris dataset, where Logistic Regression, SVM, and Decision Tree achieved 100% accuracy.
